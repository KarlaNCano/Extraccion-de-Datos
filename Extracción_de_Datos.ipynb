**FILTROS BASICOS PARA VARIABLES CUALITATIVAS **

```python
#El filtro 1 "filtro por objeto" fue aplicado en la columna "continent" para obtener los registros que solo contengan la categoria "Asia"
filtro1=df[df["continent"]=="Asia"]
filtro1
```

```python
#El filtro 2 "filtro por filas" fue aplicado en las filas 10 a 16 no consecutivamente
#filtro2=df.iloc[300:302,:]
filtro2=df.iloc[ 10:16 , : ] #filas no consecutivas
filtro2
```

```python
#El filtro 3 "filtro por columnas" fue aplicado en las columnas 1, 4 y 6 no consecutivamente
#filtro3=df.iloc[:,4:7]
filtro3=df.iloc[:,[1,4,6]] #columnas no consecutivas
filtro3
```

```python
df= pd.read_csv("CountryTable.csv", index_col=2)
df
```

```python
#El filtro 4 "filtro de filas vs columnas" fue aplicado en la columna "population" para obtener los registros que solo contengan la categoria "Asia" o "Africa"

#Paso 2: Aplicamos Filtro de filas vs columnas
filtro4=df.loc[["Asia","Africa"], ["population"]]
filtro4
```

```python
#El filtro 5 "filtro por cabecera" fue aplicado en la cabecera
filtro5=df.head(10)
filtro5
```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
#El filtro 6 "filtro por comparacion" fue aplicado en la columna "population" para obtener los registros que solo contengan la categoria menor a "2006"
filtro6=df[df["population"]<2006]
filtro6
```

**FILTROS BASICOS CON CONECTORES LOGICOS**

```python
#El filtro 7 "filtro Y" fue aplicado en la columna "population" para obtener los registros mayores a "100000" y "name" igual a "Aruba"
filtro7=df[(df["population"]>100000) & (df["name"]=="Aruba")]
filtro7
```

```python
#El filtro 8 "filtro O" fue aplicado en la columna "population" para obtener los registros que solo contengan la categoria mayor a "100000" o "name" igual a "Aruba"
filtro8=df[(df["population"]>100000) | (df["name"]=="Aruba")]
filtro8
```

```python
#El filtro 9 "filtro NOT" fue aplicado en la columna "population" para no obtener los registros que contengan la categoria mayor o igual "100000" y tampoco "name" igual a "Aruba"
filtro9=df[~(df["population"]>=100000) & ~(df["name"]=="Aruba")]
filtro9
```

```python
#Convertir archivo filtrado a CSV
filtro9.to_csv("Entregable1.csv")
```
